```python

# Extract day, month, and year from Date_of_Journey to create numerical features for modeling
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df.info()
#Converting date,Month and Year data type from object to int

df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
#Droping Date_of_Jounery column
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time']= df['Arrival_Time'].apply(lambda x:x.split(' ')[0]) #removing the char after mins
```

```python
#Separte Arrival time into Hour and Min

df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
df.info()

#Changing datatype to int

df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_min'] = df['Arrival_min'].astype(int)
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
#Separte Dep_time time into Hour and Min

df['Dep_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_min'] = df['Dep_Time'].str.split(':').str[1]

df.head()
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.info()

#Converting to int datatype

df['Dep_hour'] = df['Dep_hour'].astype(int)
df['Dep_min'] = df['Dep_min'].astype(int)
```

```python
#Changing the data to numerical and add 1 (mode value) for null column
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3,'4 stops':4,np.nan:1})
```

```python
df.head(2)
```

```python
#Droping the route columns, 
df.drop('Route',axis=1,inplace=True)
```

```python
#Getting hours and mins separte for duration

df['Duration_hour'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_min'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df.head()

```

```python
#Completing the null values

df['Duration_hour'].unique()
df['Duration_min'].unique()

#Removing the row with 5m in hours.

#Replacing nan duration_min to 0
df['Duration_min'] = df['Duration_min'].fillna(0)
```

```python
df.loc[df['Duration_hour'] == '5m', 'Duration_hour'] = 0
df.loc[df['Duration_hour'] == 0, 'Duration_min'] = 5

# One record contained a duration of '5m' only.
# Interpreted as 0 hours and 5 minutes.
```

```python

#Convert Durtaion_hour and Duration_min to int
df['Duration_hour'] = df['Duration_hour'].astype(int)
df['Duration_min'] = df['Duration_min'].astype(int)
```

```python
df.info()
```

```python
#Droping Duration

df.drop('Duration', axis=1,inplace=True)
```

```python
#Converting categoury variables into numerical using OneHotEncoder

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()
encoder_array = encoder.fit_transform(df[['Airline','Source','Destination','Additional_Info']]).toarray()
encoder_df = pd.DataFrame(encoder_array, columns=encoder.get_feature_names_out(['Airline','Source','Destination','Additional_Info']))

final_df = pd.concat([df[['Airline','Source','Destination','Additional_Info']],encoder_df], axis=1)

#print(final_df.head())

df = pd.concat([df,encoder_df],axis=1)




```

```python
#Droping Airline, Souce, Destination, Additional_Info

df.drop('Source', axis=1, inplace=True)
df.drop('Destination', axis=1, inplace=True)
df.drop('Airline', axis=1, inplace=True)
df.drop('Additional_Info', axis=1, inplace=True)
```
